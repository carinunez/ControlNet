```python
class ControlBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1),
            nn.ReLU()
        )

    def forward(self, x):
        return self.conv(x)

class ControlNetUNet(nn.Module):
    def __init__(self, unet, control_channels=3):
        super().__init__()
        self.unet = unet
        # Ejemplo: un bloque para la primera capa
        self.control_block1 = ControlBlock(control_channels, 320)
        # Puedes replicar esto para múltiples niveles

    def forward(self, x, timesteps, context, cond_image):
        ctrl1 = self.control_block1(cond_image)

        # Aquí deberías modificar la forward del U-Net para inyectar ctrl1
        # en la primera capa de downsampling, por ejemplo
        # Se recomienda congelar self.unet y solo entrenar los control_blocks

        return self.unet(x, timesteps, context)
```

```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]

image.save("astronaut_rides_horse.png")
```
